```python
import math

def Mystery(n):
    
    x = 0
    while n>=1:
        x += n % 10
        n = math.floor(n/10)
    return x

def testMystery():
    print("n | output \n------|-----|")
    for n in range(0, 1000, 1):
        if n%20==0:
            output = Mystery(n)
            print(f"{n:.0f} | {output:.0f}")

testMystery()
```

```python
def GuessWhat(A):
    
    n = len(A)
    Aux = []
    B = []
    for i in range(0, n):
        Aux.append(0)
        B.append(-1)
    
    for i in range(0, n-1, 1):
        for j in range(i+1, n, 1):
            if A[i]<A[j]:
                Aux[j] = Aux[j] + 1
            else:
                Aux[i] = Aux[i] + 1
            
            if (i==0 and j==4):
                print("6.1\t", Aux)
    
    print("6.2\t", Aux)
    
    for i in range(0, n):
        
        B[Aux[i]] = A[i]
    
    print("6.3\t", B)
    return B

def test_6_1():
    A = [32,27,94,20,53,11]
    B = GuessWhat(A)
    print(B)

test_6_1()
```
